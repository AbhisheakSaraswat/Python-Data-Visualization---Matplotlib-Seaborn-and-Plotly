```python
sales = [100,200,300,400,500]
quantity = [10,20,30,40,50]
profit = [90,160,210,280,400]

plt.style.use("fivethirtyeight")

fig, ax = plt.subplots(figsize = (10,4))
ax.bar(np.arange(len(sales)), sales, align='center')

# plot the line chart for quantity
ax.plot(np.arange(len(quantity)), quantity, color = 'red', linestyle = 'dashed')

# plot the line chart for profit
ax.plot(np.arange(len(profit)), profit, color = 'orange', linestyle = 'dotted')

# Adding labels and annotations
ax.set_xlabel('Month')
ax.set_ylabel('Value')
ax.set_xticks(np.arange(len(sales)))
ax.set_xticklabels(['Jan','Feb','Mar','Apr','May'])
ax.legend(['Quantity','Profit'])


# Adding a data labels
for i, v in enumerate(sales):
    ax.text(i, v, str(v), ha = 'center',va = 'bottom')

plt.show()
```

```python
target = 1000
sales = [920,900,1100,1050,890,1090,950,1020]
plt.plot([target] * len(sales), label = 'Target')
plt.plot(sales, label = 'Sales')
plt.legend()
plt.xlabel("Month")
plt.ylabel('Amount')
plt.title("Target vs Sales")
plt.show()
```

```python
month = lc3.Month
unitSold = lc3.UnitsSold
total_transaction = lc3.TotalTransaction
 # create figure and axis object
fig, ax = plt.subplots(figsize = (20,7))


plt.style.use('seaborn-dark')
ax.bar(month, total_transaction, align='center', color = 'darkblue', edgecolor = 'white',
      label = 'Total Transaction')

ax.set_ylabel('Total Transaction', color = 'darkblue')
ax.tick_params(axis = 'y', labelcolor = 'darkblue')

# Change Axis :

ax1 = ax.twinx()

ax1.plot(month, unitSold, color = 'red', linewidth = 4, label = "Units Sold")
ax1.set_ylabel('Units Sold', color = 'red')
ax1.tick_params(axis = 'y', labelcolor = 'red')

for i , v in enumerate(unitSold):
    ax1.text(i, v, str(v), color = 'red', ha = 'center', va = "bottom", fontsize = 'x-large')
    
ax.legend(loc = 'upper left')
ax1.legend(loc = 'upper right')
plt.show()
```
